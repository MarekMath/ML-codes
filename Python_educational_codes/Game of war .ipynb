```python
import random
```

```python
suit = ["Hearts", "Diamonds", "Clubs", "Spades"]
kind = ["Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace"]
values = {"Two":2,"Three":3,"Four":4,"Five":5,"Six":6,"Seven":7,"Eight":8,"Nine":9,"Ten":10,"Jack":11,"Queen":12,"King":13,"Ace":14}
```

```python
class card:
    
    def __init__(self, suit, kind):
        
        self.suit = suit
        self.kind = kind
        self.value = values[kind]
```

```python
class deck:
    
    def __init__(self, color):
        
        card_list = []
        self.card_list = card_list

        if color == "Red":
            for s in suit[:2]:
                for k in kind:
                    card_list.append(card(s,k))
                          
        if color == "Black":
            for s in suit[2:]:
                for k in kind:
                    card_list.append(card(s,k))
                    
        random.shuffle(card_list)

    def win_card(self,cards):
        self.card_list.extend(cards)
        
    def play_card(self):
        return self.card_list.pop(0)
```

```python
class human_deck(deck):
    
    def __init__(self):
        self.deck = deck("Red")
        
    def place_on_top(self):
        self.deck.card_list.insert(0, self.deck.card_list.pop()) 
```

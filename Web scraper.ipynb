```python
def scraper(soup):
        
    price = []
    street = []
    address = []
    location = []
    beds = []
    f_baths = []
    h_baths = []
    area = []
    lot_size = []
    l = []
    
    n = 0
    
    while len(soup.find_all("div",attrs={"class":"propertyRow"})) > 0 or price == []:
           
        r = requests.get("http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s=%s.html" % n, headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
        soup = BeautifulSoup(r.content,'html.parser')
        n += 10
    
        for i in soup.find_all("div",attrs={"class":"propertyRow"}):

            if i.find_all("h4",attrs={"class":"propPrice"}) == []:
                price.append('None')
            else: 
                price.append(i.find_all("h4",attrs={"class":"propPrice"})[0].text.strip())

            if i.find_all("span",attrs={"class":"propAddressCollapse"}) == []:
                street.append('None')
            else: 
                street.append(i.find_all("span",attrs={"class":"propAddressCollapse"})[0].text.strip())

            if i.find_all("span",attrs={"class":"propAddressCollapse"}) == []:
                address.append('None')
            else: 
                address.append(i.find_all("span",attrs={"class":"propAddressCollapse"})[1].text.strip().split(',')[0])

            if i.find_all("span",attrs={"class":"propAddressCollapse"}) == []:
                location.append('None')
            else: 
                location.append(i.find_all("span",attrs={"class":"propAddressCollapse"})[1].text.strip().split(',')[1])

            if i.find_all("span",attrs={"class":"infoBed"}) == []:
                beds.append('None')
            else: 
                beds.append(i.find_all("span",attrs={"class":"infoBed"})[0].find("b").text.strip())

            if i.find_all("span",attrs={"class":"infoValueFullBath"}) == []:
                f_baths.append('None')
            else: 
                f_baths.append(i.find_all("span",attrs={"class":"infoValueFullBath"})[0].find("b").text.strip())

            if i.find_all("span",attrs={"class":"infoValueHalfBath"}) == []:
                h_baths.append('None')
            else: 
                h_baths.append(i.find_all("span",attrs={"class":"infoValueHalfBath"})[0].find("b").text.strip())

            if i.find_all("span",attrs={"class":"infoSqFt"}) == []:
                area.append('None')
            else: 
                area.append(i.find_all("span",attrs={"class":"infoSqFt"})[0].text.strip())     

    #         for x in i.find_all("div",attrs={"class":"columnGroup"}):
    #             for atr1,atr2 in zip(x.find_all("span",attrs={"class":"featureGroup"}), x.find_all("span",attrs={"class":"featureName"})):
    #                 if 'Lot Size' in atr1.text:
    #                     lot_size.append(atr2.text) 

    #     print(lot_size)

            

    property_data = pd.DataFrame({  'price' : price,
                                    'street' : street,
                                    'address' : address,
                                    'location' : location,
                                    'beds' : beds,
                                    'f_baths' : f_baths,
                                    'h_baths' : h_baths,
                                    'area' : area,
#                                     'Lot size': lot_size
                                 })
    return property_data

scraper(soup)
```
